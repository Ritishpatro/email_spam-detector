```python
import pandas as pd
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

try:
    movies_df = pd.read_csv('movies.csv')
    print("Successfully loaded the movies data.")
except FileNotFoundError:
    print("Error: 'movies.csv' not found. Please make sure the file is in the same directory.")
    exit()

print("\nFirst 5 rows of the DataFrame:")
print(movies_df.head())

print("\nDataFrame information:")
movies_df.info()

print("\nShape of the DataFrame:", movies_df.shape)


def extract_genres(json_string):
    genres_list = []
    try:
        json_data = json.loads(json_string)
        for genre_dict in json_data:
            genres_list.append(genre_dict['name'])
    except:
        pass
    return " ".join(genres_list)

movies_df['cleaned_genres'] = movies_df['genres'].apply(extract_genres)

print("\nDataFrame with cleaned genres:")
print(movies_df[['title', 'genres', 'cleaned_genres']].head())


tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(movies_df['cleaned_genres'])

print("\nShape of the TF-IDF matrix (movies x genres):", tfidf_matrix.shape)


cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

print("\nShape of the cosine similarity matrix:", cosine_sim.shape)


indices = pd.Series(movies_df.index, index=movies_df['title']).drop_duplicates()

def get_recommendations(title, cosine_sim=cosine_sim, movies_df=movies_df, indices=indices):
    if title not in indices:
        return f"Movie '{title}' not found in the dataset. Please check the title."

    idx = indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return movies_df['title'].iloc[movie_indices]

print("\nRecommendations for 'The Dark Knight Rises':")
print(get_recommendations('The Dark Knight Rises'))

print("\nRecommendations for 'Iron Man':")
print(get_recommendations('Iron Man'))
```
